```python
import pandas as pd

df = pd.read_csv('earnings_train.csv')

# Data Quality - Part 1
# Check data types
print("Data Types:")
print(df.dtypes)
print("\n")

# Check for missing data
print("Missing Data Count:")
print(df.isnull().sum())
print("\n")

# Percentage of missing data
print("Percentage Missing Data:")
print((df.isnull().sum() / len(df)) * 100)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('earnings_train.csv')

# Range - Part 1
# List categorical columns
categorical_cols = ['DISTRICT_TYPE', 'DISTRICT_NAME', 'DEMO_CATEGORY', 
                    'STUDENT_POPULATION', 'AWARD_CATEGORY', 'ACADEMIC_YEAR']

# Show unique values for each
for col in categorical_cols:
    print(f"\n{col}:")
    print(df[col].value_counts())

# Get numeric columns
numeric_cols = ['DISTRICT_CODE', 'WAGE_YEAR1', 'WAGE_YEAR2', 'WAGE_YEAR3', 'WAGE_YEAR4']

# Show range and statistics
print(df[numeric_cols].describe())

# Check distribution with histograms
df[numeric_cols].hist(figsize=(15, 10), bins=30)
plt.tight_layout()
plt.savefig('numeric_distributions.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('earnings_train.csv')

# Semantics - Part 1
# Correlation between wage years
wage_cols = ['WAGE_YEAR1', 'WAGE_YEAR2', 'WAGE_YEAR3', 'WAGE_YEAR4']
correlation = df[wage_cols].corr()
print(correlation)

# Visualize correlation
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Between Wage Years')
plt.savefig('wage_correlation.png', dpi=300, bbox_inches='tight')
plt.show()

# Compare wages by demographics
plt.figure(figsize=(12, 6))
df.groupby('AWARD_CATEGORY')['WAGE_YEAR4'].mean().sort_values().plot(kind='barh')
plt.title('Average Year 4 Wage by Award Type')
plt.xlabel('Average Wage ($)')
plt.tight_layout()
plt.savefig('wage_by_award.png', dpi=300, bbox_inches='tight')
plt.show()
```
